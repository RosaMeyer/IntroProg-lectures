```python
import numpy as np
import matplotlib.pyplot as plt

# Range of X values
X = np.linspace(0, 4, 1000)

# Values of beta to plot
betas = [0.2, 0.5, 0.8]

# Parameter A
A = 1

plt.figure(figsize=(8, 6))

for beta in betas:
    # Compute the corresponding values of I^H for each beta
    IH = A * X**beta
    
    # Plot the production function
    plt.plot(X, IH, label=f' = {beta}')

# Add labels and title
plt.xlabel('Composite Input (X)')
plt.ylabel('Quantity of Housing Produced (I^H)')
plt.title('Production Function: I^H = A*X^')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the parameters
alpha = 0.5  # Weight of housing in utility function
C = np.linspace(0, 10, 100)  # Consumption range
H = np.linspace(0, 10, 100)  # Housing range

# Create a meshgrid for consumption and housing
C, H = np.meshgrid(C, H)

# Calculate utility using the Cobb-Douglas utility function
U = H**alpha * C**(1 - alpha)

# Create the 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot surface
ax.plot_surface(C, H, U, cmap='viridis')

# Add labels and title
ax.set_xlabel('Consumption (C)')
ax.set_ylabel('Housing Stock (H)')
ax.set_zlabel('Utility (U)')
ax.set_title('Utility as a Function of Consumption and Housing')

# Show plot
plt.show()
```
